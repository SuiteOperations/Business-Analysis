```python
# ADI method

def adi(x):
    '''
    x is an array of demand quantities. The frequency must be equal units with 0 or blanks for no pulls
    '''
    
    # Setup
    x = np.array(x)
    demand_array = np.zeros(len(x))
    
    # Loop for boolean demand
    for i in range(1, len(x)):
        
        if x[i] == 0:
            demand_array[i] = 0
        else:
            demand_array[i] = 1
            
    # ADI calculation
    adi = len(x) / len(x[x > 0])
    
    return adi
```
