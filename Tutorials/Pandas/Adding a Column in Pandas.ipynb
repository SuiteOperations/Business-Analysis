```python
import numpy as np
import pandas as pd
```

```python
# import sample data
data = pd.read_excel('./data/pandas_add_column.xlsx')
    
```

```python
data
```

```python
# create additional column
## as list
supplier_list = ['Whiskey', 'Xray', 'Yankee', 'Zulu']
## as dictionary
supplier_dictionary = {'28374':'Whiskey', '28394':'Yankee', '23928':'Zulu', '23305':'Xray'}
```

```python
# Method 1
## Declaring

data['Supplier'] = supplier_list
data.head()
```

```python
# Method 2
## insert()

data = data.drop(columns=['Supplier'])
data.insert(2,"Supplier",supplier_list,True)
data.head()
```

```python
# Method 3
## assign()

data = data.drop(columns=['Supplier'])

data = data.assign(Supplier = supplier_list)
data.head()
```

```python
# Method 4
## dictionary()

data = data.drop(columns=['Supplier'])

inverted_dict = dict([[v,k] for k,v in supplier_dictionary.items()])

data["Supplier"] = inverted_dict
data.head()
```

```python
# Method 5
## assign()
### Useful for transformations

data = data.drop(columns=['Supplier'])

data = data.assign(Supplier =lambda x: supplier_list)
data.head()
```

```python
data = data.assign(Peso_total =lambda x: data['Order Total']*19.5)
data.head()
```

```python
# Method 6
## Start with default value

data = data.drop(columns=['Supplier', 'Peso_total'])

data['Supplier'] = "none"

data.head()
```

```python
# Method 7
## Declaration from another column

data = data.drop(columns=['Supplier'])

data['Peso_total'] = data['Order Total'] * 19.5
data.head()
```
