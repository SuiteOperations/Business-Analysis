```python
import numpy as np
import pandas as pd
```

```python
report = pd.read_csv('../data/open_order_report.csv')
```

```python
report.head()
```

```python
## Due in the next month
report['PastDue Days'] = report['PastDue Days'].astype('int64')
report[report['PastDue Days'] >= -30]
```

```python
## Total by Open PO
report['total'] = report['Open Quantity'] * report['Net price']
report.groupby('Purchasing Document')[['total']].sum()
```

```python
## Total by open part
open_part_dollars = report.groupby('material_hash')[['total']].sum()
open_part_dollars.sort_values('total', ascending=False)
```
