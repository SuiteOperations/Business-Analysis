```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = po_mast.parse('pomast')
df.head()
```

```python
_ = plt.hist(df['puramt'])
_ = plt.xlabel('Order Amount')
_ = plt.ylabel('Count')
plt.show()
```

```python
plt.gcf().clear()
sns.set()
bins = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 1000, 10000, 100000]
_ = plt.hist(df['puramt'], bins=bins)
_ = plt.xlabel('Order Amount')
_ = plt.ylabel('Count')
plt.show()
```

```python
plt.gcf().clear()
_ = plt.hist(df['puramt'], bins=100)
_ = plt.xlabel('Order Amount')
_ = plt.ylabel('Count')
plt.show()
```

```python
plt.gcf().clear()
df_adjusted = df[(df['puramt'] >= 0)         & (df['puramt']        <= 10000)]
_ = plt.hist(df_adjusted['puramt'], bins=50)
_ = plt.xlabel('Order Amount')
_ = plt.ylabel('Count')
plt.show()
```

```python
plt.gcf().clear()
df_adjusted = df[(df['puramt'] >= 0) & (df['puramt'] <= 1000)]
_ = plt.hist(df_adjusted['puramt'], bins=50)
_ = plt.xlabel('Order Amount')
_ = plt.ylabel('Count')
plt.show()
```

```python
plt.gcf().clear()
_ = sns.swarmplot(x='company', y='puramt', data=df_adjusted)
_ = plt.xlabel('Vendor')
_ = plt.ylabel('Spend')
plt.show()
```

```python
plt.gcf().clear()
max_purchase = 10000
vendor = ['STAR STAINLESS SCREW', 'N D INDUSTRIES']
df_adjusted = df[(df['puramt'] >= 0) & (df['puramt'] <= max_purchase)]
df_adjusted_vendor = df_adjusted.loc[df['company'].isin(vendor)]
_ = sns.swarmplot(x='company', y='puramt', data=df_adjusted_vendor)
_ = plt.xlabel('Vendor')
_ = plt.ylabel('Spend')
plt.show()
```

```python
plt.gcf().clear()
vendor = ['STAR STAINLESS SCREW']
df_adjusted_vendor = df_adjusted.loc[df['company'].isin(vendor)]
x = np.sort(df_adjusted_vendor['puramt'])
y = np.arange(1, len(x)+1) / len(x)
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel('Purchase Ammount')
_ = plt.ylabel('ECDF')
plt.margins(0.02)
plt.show()
```
